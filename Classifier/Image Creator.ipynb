```python
import numpy as np
import os
import matplotlib.pyplot as plt
import requests
```

```python
r = requests.get('https://storage.cloud.google.com/quickdraw_dataset/sketchrnn/The%20Eiffel%20Tower.full.npz')
```

```python
DATA_PATH = 'Data/'
PROCESS_PATH = 'Process/'
```

```python
for file in os.listdir(DATA_PATH):
    file_path = DATA_PATH + file
    data = np.load(file_path, encoding='latin1', allow_pickle=True)
```

```python
test,train,valid = data['test'],data['train'],data['valid']
```

```python
for i,img in enumerate(train):
    fig, ax = plt.subplots()
    curx, cury = 0, 0
    
    for stroke in img:
        if stroke[2] == 0:
            x,y  = [curx, curx + stroke[0]], [cury, cury - stroke[1]]
            ax.plot(x,y)
        curx += stroke[0]
        cury -= stroke[1]
    
    
    plt.axis('off')
    plt.savefig(PROCESS_PATH + "/" + str(i) + '.png')
    plt.close()
```
